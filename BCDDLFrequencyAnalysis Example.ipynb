```python
import numpy as np
from BCDDLFrequencyAnalysis import BCDDLFrequencyAnalysis

%matplotlib inline
```

```python
data = np.loadtxt('ExampleSignals.txt')
T = data[:,1]
X = data[:,4]
Y = data[:,3]

num = np.loadtxt('TFNumerator.txt')
den = np.loadtxt('TFDenominator.txt')
```

```python
FA = BCDDLFrequencyAnalysis(T,X,Y,ZeroTime=True)
FA.plotData(tmin=0.008,tmax=None)
FA.plotCoherence(nmin=4,nmax=9)
FA.plotFRF(N=128)
FA.fitLowpass(N=128,f0=100,Q0=0.7,T0=0.0001,ffitmax=400)
FA.plotTFFit(N=128,num=num,den=den,T0=0.0001)
```
