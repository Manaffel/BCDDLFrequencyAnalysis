```python
import numpy as np
from BCDDLFrequencyAnalysis import BCDDLFrequencyAnalysis

%matplotlib inline
```

```python
data = np.loadtxt('ExampleSignals.txt')
T = data[:,0]
X = data[:,1]
Y = data[:,2]
```

```python
FA = BCDDLFrequencyAnalysis(T,X,Y,ZeroTime=True)
FA.plotData(tmin=0.008,tmax=None)
FA.plotCoherence(nmin=4,nmax=8,tmin=0.008)
FA.plotFRF(N=64,tmin=0.008,fmax=1000)
FA.fitLowpass(N=64,f0=100,Q0=0.7,T0=0.0001,tmin=0.008,fmax=1000,ffitmax=500)
```
